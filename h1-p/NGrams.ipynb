```python
import pandas as pd

df = pd.read_csv('gene.ngrams', sep=' ', names=['Count', 'NGram', 'One', 'Two', 'Three'])
```

```python
df
```

# O Tags

```python
os = df[df['Tag'] == 'O']
```

```python
os[['Token', 'Count']].sort('Count', ascending=False)
```

```python
os.sum()['Count']
```

# Most Common

```python
axes = os[['Count', 'Token']].sort('Count', ascending=False)[:30].plot(x='Token', y='Count', kind='bar')
axes.set_xlabel('Most Common O')
axes.set_ylabel('Count')
```

# I-GENE Tags

```python
i_genes = df[df['Tag'] == 'I-GENE']
```

```python
i_genes[['Token', 'Count']].sort('Count', ascending=False)
```

# Most Common

```python
axes = i_genes[['Count', 'Token']].sort('Count', ascending=False)[:30].plot(x='Token', y='Count', kind='bar')
axes.set_xlabel('Most Common I-GENE')
axes.set_ylabel('Count')
```
